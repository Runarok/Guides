## 1Ô∏è‚É£ Program 1: Find the Maximum Element in a List


```python
numbers = [5, 12, 7, 3, 9]
max_value = numbers[0]
for num in numbers:
    if num > max_value:
        max_value = num
print("Maximum number:", max_value)
```

## 4Ô∏è‚É£ Program 4: List Comprehension to Square Numbers


```python
numbers = [1, 2, 3, 4, 5]
squares = [n ** 2 for n in numbers]
print("Squares:", squares)
```

## 5Ô∏è‚É£ Program 5: Use Tuple as Dictionary Key


```python
coordinates = {}
coordinates[(0, 0)] = "Origin"
coordinates[(1, 2)] = "Point A"
coordinates[(2, 3)] = "Point B"
print("Coordinates Dictionary:", coordinates)
```

## 6Ô∏è‚É£ Program 6: Pack and Unpack a Tuple

## 7Ô∏è‚É£ Program 7: Count Frequency of Words


```python
sentence = "apple orange apple banana orange apple"
words = sentence.split()
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
print("Word Frequency:", frequency)
```

## 8Ô∏è‚É£ Program 8: Invert a Dictionary


```python
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {}
for key, value in original.items():
    inverted[value] = key
print("Inverted Dictionary:", inverted)
```

## 9Ô∏è‚É£ Program 9: Loop Through Dictionary
Print all key-value pairs from a dictionary of student scores.


```python
student_scores = {
    "Alice": 85,
    "Bob": 90,
    "Charlie": 78
}
for name, score in student_scores.items():
    print(name, "scored", score)
```

## üîü Program 10: Store Student Records as Tuples in a List, Then Use Dictionary for Lookup


```python
# List of student records (ID, Name, Department)
students = [
    (101, "Alice", "ECE"),
    (102, "Bob", "CSE"),
    (103, "Charlie", "ECE")
]

# Create dictionary for lookup by ID
student_dict = {}
for sid, name, dept in students:
    student_dict[sid] = (name, dept)

# Lookup a student by ID
search_id = 102
if search_id in student_dict:
    print("Details for ID", search_id, ":", student_dict[search_id])
else:
    print("Student not found.")
```
