```python
# Conceptual Python code for an AISceneAgent
# This version allows users to select a scene, input a creative prompt,
# and fetches original scene summaries from a predefined dictionary.
# It also fetches the API key from Colab secrets.


import google.generativeai as genai # Import the Gemini library
try:
    from google.colab import userdata # For Colab secrets
    COLAB_ENV = True
except ImportError:
    COLAB_ENV = False
    import os # For local environment variables as an alternative


# --- Storage for Original Scene Summaries ---
# In a real application, this might come from a file or database.
MOVIE_SCENE_SUMMARIES = {
    1: """
Scene 1: Present Day - Searching for Rancho
Farhan and Raju receive a call from Chatur, who claims he found Rancho.
The three begin a journey to find him, triggering flashbacks to college days.
Key Characters: Farhan, Raju, Chatur.
Setting: Present day, car journey, various locations.
Core Emotion: Anticipation, mystery, rekindled hope.
""",
    2: """
Scene 2: Arrival at ICE (Imperial College of Engineering) - Flashback
Farhan narrates how he, Raju, and Rancho met as engineering students.
Rancho stands out with his unconventional thinking and love for learning.
They meet strict Professor Viru Sahastrabuddhe (nicknamed Virus).
Key Characters: Rancho, Farhan, Raju, Virus, other students.
Setting: Imperial College of Engineering (ICE), past.
Core Emotion: Introduction, awe (for Rancho), intimidation (by Virus).
""",
    3: """
Scene 3: Rancho’s Rebellious Nature - Flashback
Rancho questions the rote-learning system and clashes with professors.
He befriends Farhan and Raju, encourages them to follow their passions (Farhan - photography, Raju - overcoming fear).
Rancho often challenges Virus, earning both respect from some and trouble.
Key Characters: Rancho, Farhan, Raju, Virus, other professors.
Setting: ICE classrooms, hostel.
Core Emotion: Rebellion, inspiration, humor, tension.
""",
    4: """
Scene 4: Introduction of Pia - Flashback
Rancho meets Pia, Virus’s daughter, at her sister Mona's wedding. He crashes it with Farhan and Raju.
He begins to develop feelings for her. Pia is a medical student.
Pia also disapproves of her father's rigid mindset and finds Rancho intriguing.
Key Characters: Rancho, Pia, Virus (briefly), Farhan, Raju, Suhas (Pia's then-fiancé).
Setting: Wedding venue.
Core Emotion: Romance, humor, curiosity.
""",
    5: """
Scene 5: Rancho vs. Chatur (Silencer) - Flashback
Chatur, a nerdy student who memorizes everything, challenges Rancho.
In a speech competition (Teacher's Day), Chatur delivers a memorized speech with altered words (thanks to Rancho's prank), embarrassing himself hilariously. This defines Chatur's rivalry with Rancho.
Key Characters: Rancho, Chatur, Virus, students, faculty.
Setting: ICE auditorium/event hall.
Core Emotion: Extreme humor, humiliation (for Chatur), cleverness (of Rancho).
""",
    6: """
Scene 6: Personal Struggles of Farhan and Raju - Flashback
Farhan wants to be a wildlife photographer, but his father wants him to be an engineer. Rancho encourages him to pursue his dream.
Raju is burdened by his family's poverty and his fear of failure, leading him to be overly superstitious. Rancho tries to build his confidence.
Key Characters: Farhan, Mr. Qureshi (Farhan's father), Raju, Raju's family, Rancho.
Setting: Farhan's home, Raju's home, ICE.
Core Emotion: Empathy, frustration, hope, internal conflict.
""",
    7: """
Scene 7: Joy Lobo’s Tragedy & Its Aftermath - Flashback
A creative and passionate senior student, Joy Lobo, is denied an extension by Virus for his innovative project (a drone). Under immense pressure and despair, Joy commits suicide.
Rancho is deeply affected and confronts Virus about the flaws and pressures of the education system that led to Joy's death. Virus remains dismissive.
This event hardens Rancho's resolve to challenge the system and deeply impacts Farhan and Raju.
Key Characters: Joy Lobo, Rancho, Virus, Farhan, Raju.
Setting: ICE (Joy's room, Virus's office, campus grounds).
Core Emotion: Tragedy, grief, anger, injustice, determination.
""",
    8: """
Scene 8: Raju’s Suicide Attempt & Recovery - Flashback
Virus intensifies pressure on Raju, threatening to expel him unless he writes a letter against Rancho or gets rusticated himself.
Torn between his family's dire situation (his sister's marriage, father's illness) and his loyalty to Rancho, Raju attempts suicide by jumping from Virus's office window.
He survives but is severely injured and paralyzed for a period. Rancho, Farhan, and Pia support him tirelessly during his recovery.
This event forces Raju to confront his fears head-on.
Key Characters: Raju, Virus, Rancho, Farhan, Pia, Raju's family.
Setting: ICE (Virus's office, hospital).
Core Emotion: Desperation, fear, sacrifice, friendship, resilience, hope.
""",
    9: """
Scene 9: Defying the Odds - Raju's Interview & Farhan's Choice - Flashback
Despite his physical state, Raju, with newfound courage and honesty (inspired by Rancho), attends a campus job interview. He gives a brutally honest and unconventional interview, impressing the panel and landing the job.
Simultaneously, Farhan, encouraged by Rancho and having seen Raju's transformation, finally confronts his father and expresses his desire to become a wildlife photographer instead of an engineer. His father, after initial resistance, eventually understands and supports him.
Key Characters: Raju, Farhan, Rancho, Pia, interview panel, Mr. Qureshi (Farhan's father).
Setting: Hospital, interview room, Farhan's home.
Core Emotion: Courage, triumph, honesty, parental understanding, pursuing passion.
""",
    10: """
Scene 10: Childbirth in the College Lab - Flashback
During a storm, Virus's elder daughter Mona (Pia's sister) goes into labor. Roads are blocked.
Pia, with Rancho’s inventive guidance via webcam and using college lab equipment (like a vacuum cleaner contraption), helps deliver the baby.
Virus witnesses Rancho's practical genius and resourcefulness, leading to a moment of acceptance and respect. He gives Rancho his prized "astronaut pen."
Key Characters: Rancho, Pia, Mona, Virus, Farhan, Raju.
Setting: ICE common room/makeshift delivery room.
Core Emotion: High tension, ingenuity, relief, emotional turning point for Virus.
""",
    11: """
Scene 11: Back to the Present — Finding Phunsukh Wangdu
Chatur leads Farhan and Raju to Ladakh to meet the legendary scientist Phunsukh Wangdu for a business deal.
They arrive at a unique, innovative school. They finally reunite with Rancho.
Pia also arrives, having left her wedding to Suhas, and proposes to Rancho.
The big twist: Rancho IS Phunsukh Wangdu. Chatur is utterly shocked and humiliated.
Rancho reaffirms his philosophy: "Pursue excellence, success will chase you."
Key Characters: Rancho, Farhan, Raju, Chatur, Pia.
Setting: Ladakh, the school.
Core Emotion: Joyful reunion, surprise, comedic shock, triumph of ideals.
""",
    12: """
Scene 12: Final Twist & Resolution - Present Day
The immediate aftermath of the Phunsukh Wangdu reveal.
Chatur, defeated, tries to get Rancho to sign a "declaration of defeat" from their old bet, which Rancho playfully does.
Farhan, Raju, and Pia celebrate with Rancho. Their search is over, and their friend has achieved true success on his own terms.
The film ends with the friends happy and together, reinforcing the movie's messages.
Key Characters: Rancho, Farhan, Raju, Chatur, Pia.
Setting: Ladakh.
Core Emotion: Happiness, satisfaction, humor, closure.
"""
    # Add more scene summaries as needed
}




class AISceneAgent:
    def __init__(self, llm_api_key=None):
        self.llm_operational = False
        if not llm_api_key:
            if COLAB_ENV:
                try:
                    llm_api_key = userdata.get('GOOGLE_API_KEY')
                    if not llm_api_key:
                        print("Warning: GOOGLE_API_KEY not found in Colab secrets or is empty.")
                except Exception as e:
                    print(f"Error fetching API key from Colab secrets: {e}")
            else:
                llm_api_key = os.getenv('GOOGLE_API_KEY')
                if not llm_api_key:
                    print("Warning: GOOGLE_API_KEY not found in environment variables or is empty.")


        if not llm_api_key:
            print("CRITICAL: API key is not available. LLM functionalities will be disabled.")
            self.model = None
        else:
            try:
                genai.configure(api_key=llm_api_key)
                self.model = genai.GenerativeModel('gemini-1.5-flash')
                print("Gemini Model initialized successfully. LLM is operational.")
                self.llm_operational = True
            except Exception as e:
                print(f"Error initializing Gemini Model: {e}")
                self.model = None
                print("LLM functionalities will be disabled due to initialization error.")
        self.altered_movie_chronology = []


    def _get_llm_response(self, prompt):
        if not self.llm_operational or not self.model:
            error_message = (
                "ERROR: LLM IS NOT OPERATIONAL.\n"
                "Output below is NOT from the LLM.\n"
                "Please check your API key (ensure 'GOOGLE_API_KEY' is set in Colab Secrets)\n"
                "and ensure the Gemini model initialized correctly (see logs above).\n"
                f"\n--- Original Prompt Start (for context) ---\n{prompt[:300]}...\n--- Original Prompt End ---"
            )
            print(error_message)
            return error_message
        try:
            print(f"\n--- Sending Prompt to Gemini API ---\n{prompt[:300]}...\n--- End of Truncated Prompt ---")
            response = self.model.generate_content(prompt)
            if hasattr(response, 'text') and response.text:
                return response.text
            elif response.parts:
                 return response.parts[0].text
            elif response.candidates and response.candidates[0].content.parts:
                 return response.candidates[0].content.parts[0].text
            else:
                print("Warning: LLM response structure not as expected or empty.")
                return "Error: LLM did not return a valid text response, despite being operational."
        except Exception as e:
            print(f"Error during LLM call: {e}")
            return f"Error during LLM call: {e}. LLM was operational but failed during generation."


    def add_to_chronology(self, scene_summary):
        self.altered_movie_chronology.append(scene_summary)


    def get_chronology_context(self):
        if not self.altered_movie_chronology:
            return "This is the first scene being altered."
        context = "Context from previously altered scenes:\n"
        for i, summary in enumerate(self.altered_movie_chronology):
            context += f"{i+1}. {summary}\n"
        return context


    def generate_altered_scene(self, original_scene_summary, user_creative_prompt, scene_number, scene_title):
        previous_scenes_context = self.get_chronology_context()
        prompt = f"""
        You are an AI assistant tasked with creatively altering scenes from the movie "3 Idiots".
        Your goal is to generate a new version of a scene based on a user's creative idea,
        making it different from the original but still thematically related to the movie and its characters.
        The altered scenes should be able to be combined to form a new, coherent version of the movie.


        CONTEXT FROM PREVIOUSLY ALTERED SCENES:
        {previous_scenes_context}


        CURRENT SCENE TO ALTER:
        Scene Number: {scene_number}
        Original Scene Title: "{scene_title}"
        Original Scene Summary:
        {original_scene_summary}


        USER'S CREATIVE PROMPT FOR THIS SCENE:
        "{user_creative_prompt}"


        TASK:
        Interpret the USER'S CREATIVE PROMPT. Based on this prompt, the original scene summary,
        and the context from previously altered scenes (if any), generate a new narrative for this scene.
        Expand creatively on the user's prompt. The narrative should be detailed, describing actions,
        key dialogues (if applicable), setting, and emotional tone.

        Your output MUST include these clearly marked sections:
        **Altered Scene Narrative:**
        [Your generated narrative here, creatively interpreting and expanding on the user's prompt within the scene's context]


        **Key Changes Made (inspired by user's prompt):**
        - [Describe the primary change stemming from the user's prompt]
        - [Describe any secondary significant changes]


        **Continuity Notes for Next Scene:**
        - [How this user-prompted alteration might affect the next logical scene in the story or character arcs.]


        Maintain the spirit of "3 Idiots" (mix of humor, emotion, social commentary) unless the
        user's creative prompt suggests a significant tonal shift. Ensure characters behave in a way that
        is plausible given their personalities and the altered circumstances suggested by the user.
        If the user's prompt is very brief, use your creativity to flesh it out into a compelling scene change.
        """
        altered_scene_narrative = self._get_llm_response(prompt)
        summary_for_chronology = f"Altered Scene {scene_number} ('{scene_title}') based on user prompt: '{user_creative_prompt[:70]}...'."
        if not self.llm_operational:
            summary_for_chronology += " (LLM NOT OPERATIONAL)"
        self.add_to_chronology(summary_for_chronology)
        return altered_scene_narrative


    def generate_extra_scene(self, context_description, user_creative_prompt_for_new_scene, new_scene_number, new_scene_title):
        previous_scenes_context = self.get_chronology_context()
        prompt = f"""
        You are an AI assistant tasked with creating an entirely NEW scene for an altered version
        of the movie "3 Idiots", based on a user's creative idea. This scene should logically fit
        within the established altered narrative.


        CONTEXT FROM PREVIOUSLY ALTERED SCENES:
        {previous_scenes_context}


        SITUATION LEADING TO THIS NEW SCENE:
        {context_description}


        USER'S CREATIVE PROMPT FOR THIS NEW SCENE:
        "{user_creative_prompt_for_new_scene}"


        NEW SCENE DETAILS:
        New Scene Number: {new_scene_number}
        New Scene Title: "{new_scene_title}"

        TASK:
        Interpret the USER'S CREATIVE PROMPT FOR THIS NEW SCENE. Generate a detailed narrative for this new scene.
        Describe actions, key dialogues, setting, emotional tone, and how it contributes to the overall altered story.
        Expand creatively on the user's prompt.

        Your output MUST include these clearly marked sections:
        **New Scene Narrative:**
        [Your generated narrative here]


        **Purpose of this New Scene (based on user's prompt):**
        [Explain its role in the narrative]


        **Continuity Notes for Next Scene:**
        - [How this new addition might impact subsequent events.]
        """
        new_scene_narrative = self._get_llm_response(prompt)
        summary_for_chronology = f"Extra Scene {new_scene_number} ('{new_scene_title}') based on user prompt: '{user_creative_prompt_for_new_scene[:70]}...'."
        if not self.llm_operational:
            summary_for_chronology += " (LLM NOT OPERATIONAL)"
        self.add_to_chronology(summary_for_chronology)
        return new_scene_narrative


# --- Main Interactive Loop ---
if __name__ == "__main__":
    agent = AISceneAgent()


    if not agent.llm_operational:
        print("\nAgent initialized, but LLM is NOT operational. Please check API key and logs.")
        print("You can still see the structure, but outputs will be error messages.")
    else:
        print("\nAgent and LLM initialized. LLM is operational.")

    print("\nWelcome to the '3 Idiots' AI Scene Alteration Tool!")
    print("----------------------------------------------------")


    while True:
        # Sort the keys for ordered display
        available_scenes = sorted(list(MOVIE_SCENE_SUMMARIES.keys()))
        print("\nAvailable original scene numbers with summaries:", available_scenes)
        try:
            scene_num_str = input("Enter the scene number you want to alter (or type 'quit' to exit): ")
            if scene_num_str.lower() == 'quit':
                break
            scene_to_alter = int(scene_num_str)
            if scene_to_alter not in MOVIE_SCENE_SUMMARIES:
                print(f"Error: Scene {scene_to_alter} not found in summaries. Please choose from available scenes.")
                continue
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue


        original_summary = MOVIE_SCENE_SUMMARIES[scene_to_alter]
        # Extracting a simple title from the first line of the summary for now
        original_title = original_summary.strip().split('\n')[0].replace("Scene "+str(scene_to_alter)+": ","") # Clean up title

        print(f"\n--- Original Summary for Scene {scene_to_alter}: {original_title} ---")
        print(original_summary)
        print("----------------------------------------------------")

        user_prompt = input(f"Enter your creative prompt for altering Scene {scene_to_alter}: ")
        if not user_prompt.strip():
            print("No creative prompt provided. Skipping alteration for this scene.")
            continue


        print(f"\n>>> Generating Altered Scene {scene_to_alter} based on your prompt: '{user_prompt}'")
        altered_scene_output = agent.generate_altered_scene(
            original_scene_summary=original_summary,
            user_creative_prompt=user_prompt,
            scene_number=scene_to_alter,
            scene_title=original_title # Pass the cleaned title
        )
        print(f"\n--- Output for Altered Scene {scene_to_alter} ---\n{altered_scene_output}")
        print("----------------------------------------------------")

        print("\n--- Current Altered Movie Chronology ---")
        print(agent.get_chronology_context())
        print("----------------------------------------------------")


    print("\nExiting Scene Alteration Tool. Thank you!")
```
