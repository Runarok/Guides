```python
# Simulated generator function (replace with actual model inference)
def generator(prompt, max_new_tokens=300, num_return_sequences=1):
    return [{
        "generated_text": f"Generated Scene:\n{prompt}\n\n[End of Scene]"
    }]

# Create the powerful unified narrator agent
def create_master_agent():
    return {
        "role": "Master Narrator",
        "goal": "Narrate scenes with emotional depth, clarity, and narrative cohesion.",
        "backstory": """You are a powerful storytelling AI trained to weave complex, emotionally compelling scenes.
Each scene may involve flashbacks, emotional transitions, character development, or dramatic events.
You adapt each scene based on user-requested changes while maintaining narrative consistency and tone.""",
        "verbose": True
    }

# Generate narrative based on scene input and desired changes
def generate_scene(master_agent, scene_number, original_scene_summary, changes):
    prompt = f"""
{master_agent['backstory']}
Your Role: {master_agent['role']}
Goal: {master_agent['goal']}

Scene {scene_number} Summary:
{original_scene_summary}

Requested Changes:
{changes}

Now, rewrite the scene integrating the above changes while maintaining a rich, emotional, and coherent narrative style.
"""
    response = generator(prompt, max_new_tokens=300)
    return response[0]['generated_text']
```
